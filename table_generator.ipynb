```python
from collections import defaultdict
import os

import numpy as np
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_files(data_dir):
    hps = []
    files = os.listdir(data_dir)
    files = [f for f in files if f.endswith('.csv')]
    for i,f in enumerate(files):
        file_path = os.path.join(data_dir, f)
        tmp= f.split(':')
        hp = {}
        for i,param in enumerate(f.split('_')):
            vals= param.split(':')
            if i !=0:
                if i!= len(f.split('_'))-1:
                    hp[vals[0]]=eval(vals[1])
                else:
                    hp[vals[0]]=eval(vals[1].split('.csv')[0])
                
        #hp['lr']=float(tmp[1].split('_')[0])
        #hp['mom']=float(tmp[2].split('_')[0])
        #hp['epochs']=float(tmp[3].split('_')[0])
        #hp['bs']=float(tmp[4].split('_')[0])
        #hp['seed']=int(tmp[5].split('.')[0])
        
        data=pd.read_csv(file_path)
        hp['Train_Loss_AVG'] = data.iloc[:,1].values
        hp['Test_Accuracy'] = data.iloc[:,2].values
        
       # for i,acc in enumerate(data.iloc[:,2].values):
       #     hp[f'accuracy_ep{i}'] = acc 
        hps.append(hp)
        # with open(file_path, 'r') as inf:
        #     cdata = json.load(inf)
        # data.update(cdata['user_data'])
        # directories[f]=[]
        # for user, _ in cdata['user_data'].items():
        #     directories[f].append(user)
    
    return hps 

files = read_files('Results')
```
